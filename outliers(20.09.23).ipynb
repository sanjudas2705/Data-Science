```python
# outlayer detection using iqr
import pandas as pd
import numpy as np
df=pd.read_csv("heights1.csv")
df
```

```python
Q1=df.height.quantile(0.25,interpolation = 'midpoint')
Q3=df.height.quantile(0.75,interpolation = 'midpoint')
print(Q1)
print(Q3)
```

```python
#IQR
iqr=Q3-Q1
print(iqr)
```

```python
lower_limit=Q1 -1.5*iqr
upper_limit=Q3+1.5*iqr
print(lower_limit)
print(upper_limit)
```

```python
# here are the outlayers
df[(df.height<lower_limit) | (df.height>upper_limit)]
```

```python
# remove outliers
df_no_outliers=df[(df.height>lower_limit)&(df.height<upper_limit)]
df_no_outliers
```
